###1. Tạo mảng 20 số nguyên ngẫu nhiên kích thước 3 x 3 có giá trị trong khoảng [0, 20]. Viết chương trình kiểm tra tất cả các phần tử trong mảng khác 0 (tìm hiểu hàm all của numpy)

```python
import numpy as np

array = np.random.randint(0, 21, size=(3, 3))
print("Mảng ngẫu nhiên 3x3:")
print(array)
check_nonzero = np.all(array != 0)

print(f"\nTất cả các phần tử trong mảng có khác 0 không? {check_nonzero}")
```

###3. Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater, numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.

```python
import numpy as np

# Khởi tạo mảng 1 chiều
a1 = np.array([1, 5, 10])
b1 = np.array([2, 5, 8])

print("Mảng 1D a:", a1)
print("Mảng 1D b:", b1)

print("\nSo sánh 1 chiều:")
print("a > b  (greater):      ", np.greater(a1, b1))
print("a >= b (greater_equal):", np.greater_equal(a1, b1))
print("a < b  (less):         ", np.less(a1, b1))
print("a <= b (less_equal):   ", np.less_equal(a1, b1))
```

```python
# Khởi tạo mảng 2 chiều
a2 = np.array([[1, 2], [3, 4]])
b2 = np.array([[0, 2], [5, 4]])

print("Mảng 2D a:\n", a2)
print("Mảng 2D b:\n", b2)

print("\nSo sánh 2 chiều (greater_equal):")
print(np.greater_equal(a2, b2))

print("\nSo sánh 2 chiều (less):")
print(np.less(a2, b2))
```

###4. Viết chương trình tạo mảng gồm 10 số 1, 10 số 0 và 10 số 5

###8. Viết chương trình tạo mảng 20 phần tử trong khoảng [0, 20] và đổi dấu những số nằm trong khoảng [9, 15]

```python
import numpy as np

array_8 = np.linspace(0, 20, 20)

print("Mảng ban đầu:")
print(array_8)

array_8[(array_8 >= 9) & (array_8 <= 15)] *= -1

print("\nMảng sau khi đổi dấu các phần tử trong khoảng [9, 15]:")
print(array_8)
```

###11. Viết chương trình tạo ma trận 5 x 5, các phần tử ở đường chéo chính là 1, 2, 3, 4, 5. (tìm hiểu hàm diag của numpy)

###12. Viết chương trình tạo mảng 3 x 3 x 3 và tính tổng theo mỗi dòng, mỗi cột. (tìm hiểu hàm sum của numpy)

```python
import numpy as np

array_3d = np.random.randint(1, 10, size=(3, 3, 3))
print("Mảng 3x3x3:")
print(array_3d)

sum_rows = np.sum(array_3d, axis=2)

sum_cols = np.sum(array_3d, axis=1)

print("\nTổng theo mỗi dòng (mỗi hàng là kết quả của một lát cắt 3x3):")
print(sum_rows)

print("\nTổng theo mỗi cột (mỗi hàng là kết quả của một lát cắt 3x3):")
print(sum_cols)
```

### 14. Tạo ma trận A= 4 x 3 và vector y gồm 3 phần tử ngẫu nhiên. Thêm vector y vào từng dòng của ma trận A

### 15. Viết chương trình Pandas để cộng, trừ, nhân, chia 2 Series : [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]

```python
import pandas as pd

s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 10])

print("Series 1:\n", s1)
print("\nSeries 2:\n", s2)

print("\nCộng:")
print(s1 + s2)

print("\nTrừ:")
print(s1 - s2)

print("\nNhân:")
print(s1 * s2)

print("\nChia:")
print(s1 / s2)
```
