## TIỀN XỬ LÝ DỮ LIỆU TIÊU THỤ THUỐC LÁ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load dữ liệu
url = 'https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/Ecdat/Cigarette.csv'
df_cig = pd.read_csv(url).iloc[:, 1:]

print("---Kiểm tra thông tin thiếu ---")
print(f"Tổng số dòng: {len(df_cig)}")
missing_data = df_cig.isnull().sum()
missing_percent = (missing_data / len(df_cig)) * 100

missing_df = pd.DataFrame({'Missing Values': missing_data, 'Percentage (%)': missing_percent})
display(missing_df)
```

```python
print("---Xử lý giá trị thiếu ---")
if df_cig.isnull().values.any():
    num_cols = df_cig.select_dtypes(include=[np.number]).columns
    for col in num_cols:
        df_cig[col] = df_cig.groupby('state')[col].transform(lambda x: x.fillna(x.mean()))

    df_cig.dropna(inplace=True)

print("Số lượng giá trị rỗng sau xử lý:", df_cig.isnull().sum().sum())
```

```python
print("---Phát hiện và Xử lý giá trị ngoại lai ---")

# Trực quan hóa bằng Boxplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df_cig, y='packpc', ax=axes[0])
axes[0].set_title('Boxplot Doanh số (packpc)')
sns.boxplot(data=df_cig, y='avgprs', ax=axes[1])
axes[1].set_title('Boxplot Giá (avgprs)')
plt.show()

# Hàm xử lý Outlier bằng IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Số lượng Outliers trong {column}: {len(outliers)}")
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_cig = remove_outliers(df_cig, 'packpc')
df_cig = remove_outliers(df_cig, 'avgprs')
print(f"Số dòng còn lại sau khi xóa outliers: {len(df_cig)}")
```

```python
print("--- Cân bằng và Chuẩn hóa dữ liệu ---")

# Kiểm tra tính cân bằng
state_counts = df_cig['state'].value_counts()
print("Số lượng năm quan sát trung bình mỗi bang:", state_counts.mean())

# Chuẩn hóa dữ liệu bằng StandardScaler
scaler = StandardScaler()
cols_to_scale = ['pop', 'avgprs', 'packpc', 'income', 'tax', 'taxs']
df_cig[cols_to_scale] = scaler.fit_transform(df_cig[cols_to_scale])

# Xuất kết quả
df_cig.to_csv('cleaned_cigarette_data.csv', index=False)
print("Đã lưu file: cleaned_cigarette_data.csv")
display(df_cig.head())
```

```python
# 3. Xử lý dữ liệu thiếu cho cột 'DT'
print(f"Giá trị duy nhất trong DT trước khi xử lý: {df['DT'].unique()}")
df['DT'] = df['DT'].fillna(0)
print(f"Giá trị duy nhất trong DT sau khi xử lý: {df['DT'].unique()}")

# 4. Xử lý dữ liệu thiếu cho 'T1' bằng phương pháp giá trị trung bình
print(f"Số lượng giá trị thiếu trong T1: {df['T1'].isnull().sum()}")
df['T1'] = df['T1'].fillna(df['T1'].mean())

# 5. Tự động xử lý dữ liệu thiếu cho các cột điểm số còn lại bằng Mean
score_cols = ['L1','H1','S1','V1','X1','D1','N1','T2','L2','H2','S2','V2','X2','D2','N2','T6','L6','H6','S6','V6','X6','D6','N6']
for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# 6. Tạo biến điểm trung bình (TBM1, TBM2, TBM3)
# Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
def calculate_tbm(row, suffix):
    t, l, h, s, v, x, d, n = [row[f'{m}{suffix}'] for m in ['T','L','H','S','V','X','D','N']]
    return (t*2 + l + h + s + v*2 + x + d + n) / 10

df['TBM1'] = df.apply(lambda r: calculate_tbm(r, '1'), axis=1)
df['TBM2'] = df.apply(lambda r: calculate_tbm(r, '2'), axis=1)
df['TBM3'] = df.apply(lambda r: calculate_tbm(r, '6'), axis=1)

# 7. Tạo biến xếp loại (XL1, XL2, XL3) dựa trên TBM
def classify(score):
    if score < 5.0: return 'Y'
    elif score < 6.5: return 'TB'
    elif score < 8.0: return 'K'
    elif score < 9.0: return 'G'
    else: return 'XS'

for i in [1, 2, 3]:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(classify)

# 8. Chuyển đổi sang thang điểm 4 của Mỹ (US_TBM)
for i in [1, 2, 3]:
    df[f'US_TBM{i}'] = (df[f'TBM{i}'] / 10) * 4

display(df[['TBM1', 'XL1', 'US_TBM1']].head())
```

```python
# 9. Tính toán kết quả xét tuyển
def xet_tuyen(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if kt in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        score = (dh1 + dh2 + dh3) / 3

    return 1 if score >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)

# 10. Lưu trữ dữ liệu
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("Đã lưu file: processed_dulieuxettuyendaihoc.csv")
display(df.head())
```

### Phần 1: Thống kê dữ liệu

```python
import pandas as pd
import numpy as np

# Tải dữ liệu trực tiếp từ URL để đảm bảo biến df được khởi tạo thành công
url = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv'
df = pd.read_csv(url)

# Tiền xử lý dữ liệu cần thiết cho Lab 2
df['DT'] = df['DT'].fillna(0)
df['T1'] = df['T1'].fillna(df['T1'].mean())

def calculate_tbm(row, suffix):
    cols = ['T','L','H','S','V','X','D','N']
    vals = [row[f'{m}{suffix}'] for m in cols]
    return (vals[0]*2 + vals[1] + vals[2] + vals[3] + vals[4]*2 + vals[5] + vals[6] + vals[7]) / 10

for i, s in [(1, '1'), (2, '2'), (3, '6')]:
    df[f'TBM{i}'] = df.apply(lambda r: calculate_tbm(r, s), axis=1)
    df[f'US_TBM{i}'] = (df[f'TBM{i}'] / 10) * 4
    df[f'XL{i}'] = df[f'TBM{i}'].apply(lambda x: 'Y' if x<5 else ('TB' if x<6.5 else ('K' if x<8 else ('G' if x<9 else 'XS'))))

def xet_tuyen(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']
    if kt in ['A', 'A1']: score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B': score = (dh1 + dh2*2 + dh3) / 4
    else: score = (dh1 + dh2 + dh3) / 3
    return 1 if score >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)

# Hàm tính toán các đại lượng thống kê
def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)

print("Dữ liệu đã được tải và tiền xử lý thành công cho Lab 2.")
display(df.head())
```

### PHẦN 1: DATA CLEANSING & FEATURE ENGINEERING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dữ liêu
def load_data(url):
    """Loads a CSV file from a URL into a pandas DataFrame."""
    df = pd.read_csv(url)
    return df

# URL của file 23_HOMES.csv
url_homes = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/23_HOMES.csv'
df_homes = load_data(url_homes)

print("10 dòng đầu tiên của dữ liệu 23_HOMES.csv:")
display(df_homes.head(10))
```

### Xử lý dữ liệu thiếu

```python
# Thống kê số lượng giá trị thiếu trên từng cột
missing_values = df_homes.isnull().sum()
missing_percentage = (missing_values / len(df_homes)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)

print("Thống kê giá trị thiếu trên từng cột:")
display(missing_df)

# Trực quan hóa dữ liệu thiếu bằng biểu đồ Heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(df_homes.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap hiển thị dữ liệu thiếu (NaN)')
plt.show()
```

### Khai thác thông tin tổng hợp (Aggregation) - TotalSpace

```python
# 8. Tạo biến TotalSpace (Tổng không gian) bằng cách cộng các diện tích thành phần
# Giả sử Area đã tính bằng square feet, chuyển Acres sang square feet (1 acre = 43560 sq ft)
df_homes['TotalSpace'] = df_homes['Area'] + (df_homes['Acres'] * 43560)

print("5 dòng đầu tiên với TotalSpace:")
display(df_homes[['Area', 'Acres', 'TotalSpace']].head())
```

### Tạo biến định danh nhị phân (Binary Feature) - IsNew

```python
# 9. Tạo đặc trưng IsNew để xác định nhà mới hay cũ
# Nếu nhà xây sau năm 2010 thì IsNew = 1, ngược lại là 0.
df_homes['IsNew'] = (df_homes['YearBuilt'] > 2010).astype(int)

print("Phân bố của biến IsNew:")
display(df_homes['IsNew'].value_counts())
display(df_homes.head())
```

## PHẦN 2: KHAI THÁC THÔNG TIN HỮU ÍCH – EDA

### 1. Tương quan giá bán: Trực quan tỉ lệ nhà có giá cao/thấp trên từng nhóm số lượng phòng ngủ (Bedrooms).

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_homes, x='Bedrooms', hue='PriceGroup', palette='viridis')
plt.title('Phân bố PriceGroup theo số lượng phòng ngủ')
plt.xlabel('Số lượng phòng ngủ')
plt.ylabel('Số lượng nhà')
plt.legend(title='Phân khúc giá')
plt.show()
```

### 2. Phân khúc nhà: Trực quan số lượng nhà theo từng nhóm phân loại chất lượng xây dựng (sử dụng Century).

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df_homes, x='Century', palette='magma', order=sorted(df_homes['Century'].unique()))
plt.title('Số lượng nhà theo Thế kỷ xây dựng')
plt.xlabel('Thế kỷ xây dựng')
plt.ylabel('Số lượng nhà')
plt.show()
```

### 3. Phân tích đa biến: Trực quan mối quan hệ giữa giá nhà (Selling_Price) trên từng nhóm PriceGroup và năm xây dựng (YearBuilt).

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_homes, x='YearBuilt', y='Selling_Price', hue='PriceGroup', palette='viridis', alpha=0.7)
plt.title('Mối quan hệ giữa Giá bán, Năm xây dựng và Phân khúc giá')
plt.xlabel('Năm xây dựng')
plt.ylabel('Giá bán')
plt.legend(title='Phân khúc giá')
plt.grid(True)
plt.show()
```

```python
# 6. Tạo biến PriceGroup dựa trên cột Selling_Price
# Định nghĩa các ngưỡng cho PriceGroup dựa trên Selling_Price quartiles
low_price = df_homes['Selling_Price'].quantile(0.25)
medium_price = df_homes['Selling_Price'].quantile(0.50)
high_price = df_homes['Selling_Price'].quantile(0.75)

def categorize_price(price):
    if price <= low_price:
        return 'Cheap'
    elif price <= medium_price:
        return 'Affordable'
    elif price <= high_price:
        return 'Luxury'
    else:
        return 'Ultra-Luxury'

df_homes['PriceGroup'] = df_homes['Selling_Price'].apply(categorize_price)

print("Phân bố của biến PriceGroup:")
display(df_homes['PriceGroup'].value_counts())
display(df_homes.head())
```

### Tách đặc trưng (Feature Splitting) - Century từ YearBuilt (suy ra từ Age)

```python
# 7. Tách năm xây dựng (YearBuilt) để lấy ra thông tin Thế kỷ xây dựng (Century)
# Giả sử năm hiện tại là 2024 để tính YearBuilt từ Age
current_year = 2024
df_homes['YearBuilt'] = current_year - df_homes['Age']

def get_century(year):
    return f"Century {int(np.ceil(year / 100))}"

df_homes['Century'] = df_homes['YearBuilt'].apply(get_century)

print("Phân bố của biến Century:")
display(df_homes['Century'].value_counts())
display(df_homes.head())
```
