###3. Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater, numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.

```python
import numpy as np

# Khởi tạo mảng 1 chiều
a1 = np.array([1, 5, 10])
b1 = np.array([2, 5, 8])

print("Mảng 1D a:", a1)
print("Mảng 1D b:", b1)

print("\nSo sánh 1 chiều:")
print("a > b  (greater):      ", np.greater(a1, b1))
print("a >= b (greater_equal):", np.greater_equal(a1, b1))
print("a < b  (less):         ", np.less(a1, b1))
print("a <= b (less_equal):   ", np.less_equal(a1, b1))
```

```python
import pandas as pd
import numpy as np

# 2. Tải dữ liệu
url = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv'
df = pd.read_csv(url)

df.to_csv('dulieuxettuyendaihoc.csv', index=False)

print("10 dòng đầu tiên:")
display(df.head(10))
print("\n10 dòng cuối cùng:")
display(df.tail(10))
```

```python
# 3. Xử lý dữ liệu thiếu cho cột 'DT'
print(f"Giá trị duy nhất trong DT trước khi xử lý: {df['DT'].unique()}")
df['DT'] = df['DT'].fillna(0)
print(f"Giá trị duy nhất trong DT sau khi xử lý: {df['DT'].unique()}")

# 4. Xử lý dữ liệu thiếu cho 'T1' bằng phương pháp giá trị trung bình
print(f"Số lượng giá trị thiếu trong T1: {df['T1'].isnull().sum()}")
df['T1'] = df['T1'].fillna(df['T1'].mean())

# 5. Tự động xử lý dữ liệu thiếu cho các cột điểm số còn lại bằng Mean
score_cols = ['L1','H1','S1','V1','X1','D1','N1','T2','L2','H2','S2','V2','X2','D2','N2','T6','L6','H6','S6','V6','X6','D6','N6']
for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# 6. Tạo biến điểm trung bình (TBM1, TBM2, TBM3)
# Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
def calculate_tbm(row, suffix):
    t, l, h, s, v, x, d, n = [row[f'{m}{suffix}'] for m in ['T','L','H','S','V','X','D','N']]
    return (t*2 + l + h + s + v*2 + x + d + n) / 10

df['TBM1'] = df.apply(lambda r: calculate_tbm(r, '1'), axis=1)
df['TBM2'] = df.apply(lambda r: calculate_tbm(r, '2'), axis=1)
df['TBM3'] = df.apply(lambda r: calculate_tbm(r, '6'), axis=1)

# 7. Tạo biến xếp loại (XL1, XL2, XL3) dựa trên TBM
def classify(score):
    if score < 5.0: return 'Y'
    elif score < 6.5: return 'TB'
    elif score < 8.0: return 'K'
    elif score < 9.0: return 'G'
    else: return 'XS'

for i in [1, 2, 3]:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(classify)

# 8. Chuyển đổi sang thang điểm 4 của Mỹ (US_TBM)
for i in [1, 2, 3]:
    df[f'US_TBM{i}'] = (df[f'TBM{i}'] / 10) * 4

display(df[['TBM1', 'XL1', 'US_TBM1']].head())
```

```python
# 9. Tính toán kết quả xét tuyển
def xet_tuyen(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if kt in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        score = (dh1 + dh2 + dh3) / 3

    return 1 if score >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)

# 10. Lưu trữ dữ liệu
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("Đã lưu file: processed_dulieuxettuyendaihoc.csv")
display(df.head())
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Khảo sát phân phối cho biến T1
print("Mô tả độ tập trung và phân tán của T1:")
print(df['T1'].describe())

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Box-plot
sns.boxplot(y=df['T1'], ax=axes[0])
axes[0].set_title('Box-plot T1')

# Histogram
sns.histplot(df['T1'], kde=True, ax=axes[1])
axes[1].set_title('Histogram T1')

# QQ-plot
stats.probplot(df['T1'], dist="norm", plot=axes[2])
axes[2].set_title('QQ-plot T1')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Khảo sát tương quan giữa DH1 và T1
correlation = df['DH1'].corr(df['T1'])
covariance = df['DH1'].cov(df['T1'])

print(f"Correlation giữa DH1 và T1: {correlation}")
print(f"Covariance giữa DH1 và T1: {covariance}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='T1', y='DH1', data=df, hue='KV')
plt.title('Biểu đồ Scatter DH1 theo T1')
plt.show()
```
